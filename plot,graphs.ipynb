line chart

```python
import matplotlib.pyplot as plt
empl_count=[3,20,50,200,350,400]
year=[2014,2015,2016,2017,2018,2019]
#lot a line chart
plt.plot(year,empl_count,'*--g')
#set axis tittles
plt.xlabel('year')
plt.ylabel('employees')
#set chart tittle
plt.title('Employee growth at XYZ')
plt.show()
```

Bar plotting

```python
import matplotlib.pyplot as plt
sub=['python','java','cn','dbms']
students=[40,30,20,50]
plt.bar(sub,students)
plt.show()
```

Pie chart showinig intrest of 200 class students in sports

```python
from matplotlib import pyplot as plt
import numpy as np
sports=['cricket','kabbaddi','Football','chess','vallyball']
students=[50,40,60,22,40]
plt.pie(students, labels=sports,autopct=' %1.2f%%')
plt.show()
```

Histogram plot

```python
import matplotlib.pyplot as plt
x=[1,6,9,3,10,12,13,16,15,18,41,45,48,50,60,61,62,68,66,
   19,20,21,23,25,24,44,33,22,85,35,100,78,73,25,85,
   86,45]
plt.hist(x)
plt.xlabel('age')
plt.ylabel('count')
plt.title('population')
plt.show()
```

Scatter plot

```python
import matplotlib.pyplot as plt
#scatter plot
x=[2,3,4,5,6,7,]
y=[8, 8.5, 9.5, 10, 10.5, 11]
plt.scatter(x,y,label='values of x & y',color='r')
plt.title('Smart band data report')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Saving plot as image

```python
import matplotlib.pyplot as plt
#creating data
year=['2010','2002','2004','2006','2008']
production=[25,15,35,30,10]
#plotting barchart
plt.bar(year,production)
#saving the figure.
plt.savefig('output.jpg')
```
